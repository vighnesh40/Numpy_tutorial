```python
#4. Array indexing and slicing
```

```python
a = np.array([1,3,5,7,9])
print(a)
```

```python
print(a[4])
```

```python
print(a[0])
```

```python
print(a[-4])
```

```python
print(a[-1])
```

```python
# Slicing : 
# array[start:stop:step]
#stop-index,that is not included in output
a = np.array([1,3,5,7,9])
print(a)
```

```python
print(a[0:3]) # First 3 elements
```

```python
print(a[-3:]) #last 3 elements
```

```python
print(a[1:4]) # middle 3 elements
```

```python
print(a[0:4:2]) # steps
```

```python
#6.Array Stacking and Splitting
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.vstack((a,b)))
print(np.hstack((a,b)))
```

```python
c = np.array([[1,2,3],[4,5,6]])
print(c)
```

```python
split =np.hsplit(c,3)

for s in split:
    print(s)
```

```python
vsplit =np.vsplit(c,2)

for s in vsplit:
    print(s)
```

```python
#8.Mathematical Operations on Multiple Arrays
```

```python
print(np.dot(a,b)) #dot product
```

```python
print(a.T) #Transpose
```
